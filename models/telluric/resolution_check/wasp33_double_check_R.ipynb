# Resolution double-check
Resolution double check for possible factor 2 missing in FWHM as suggested by Matteo.

```python
import numpy as np
from matplotlib import pyplot as plt

data = np.loadtxt('atran15000.txt')
wav_15k, spec_15k = data[:,1], data[:,2]

data = np.loadtxt('atran30000.txt')
wav_30k, spec_30k = data[:,1], data[:,2]

plt.plot(wav_15k, spec_15k, color='k', label='R=15,000')
plt.plot(wav_30k, spec_30k, color='r', label='R=30,000')
plt.xlim(2.29,2.3)
plt.legend()
plt.xlabel('Wavelength (micron)')
plt.ylabel('Transmission')
plt.show()
```

```python
wav, spec_corr = np.load('wav_example.npy'), np.load('spec_corr_example.npy')

plt.scatter(wav,spec_corr, color='b', label='example observed spectrum', s=5)
plt.plot(wav_15k, spec_15k, color='k', label='ATRAN downloaded at R=15,000')
plt.plot(wav_30k, spec_30k, color='r', label='ATRAN downloaded at R=30,000')
plt.xlim(2.295,2.3)
plt.legend()
plt.xlabel('Wavelength (micron)')
plt.ylabel('Transmission')
plt.title('Resolution factor 2 check? -> seems lower resolution is correct')
plt.show()
```

Resolution definitely looks more to be close to the black line than the red one.
