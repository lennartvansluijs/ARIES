# Plot WASP-33 15/10/2016 output
Load and plot WASP-33 b 15/10/2016 aligned and wavelength-calibrated spectral time series.

```python
from astropy.io import fits
import os
from matplotlib import pyplot as plt
import numpy as np
```

```python
order = 23
target = 'wasp33'
date = '15102016'
```

```python
dirin = f'data/{target}/{date}/processed/spectra/alignment_with_stretching/order_{order}'
fname = 'wasp33_order_aligned.fits'
f = os.path.join(dirin, fname)
data = fits.getdata(f)
data = data[:,2:-3] # remove edges

plt.figure(figsize=(7.5,7.5))
plt.imshow(data, cmap='inferno', extent=[1, 1024, 1, data.shape[0]], aspect=10)
plt.title('Spectral time series; ARIES/MMT; WASP-33; 15/10/2016; order=23', size=15)
plt.ylabel('# frame', size=15)
plt.xlabel('# pixel', size=15)
plt.show()
```

Plot wavelength solution.

```python
dirin = f'data/{target}/{date}/processed/spectra/autowavcal'
fname = f'{target}_wavcal_order_{order}_wavsolution.txt'
f = os.path.join(dirin, fname)
pix, wav = np.loadtxt(f).T

plt.figure(figsize=(7.5,7.5))
plt.title('Wavelength solution; WASP-33; 15/10/2016; order=23',size=15)
plt.plot(pix, wav)
plt.xlabel('# pixel',size=15)
plt.ylabel('Wavelength (micron)',size=15)
plt.show()
```

Information on time stamps/velocities can be extracted from raw science frame headers.
